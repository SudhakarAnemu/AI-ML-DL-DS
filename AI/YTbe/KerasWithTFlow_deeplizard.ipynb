## Data preparation and process

```
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
from sklearn.utils import shuffle

print("Done - Import")

train_labels = []
train_samples = []

print("Done - Configuring train arrays")
```

```
for i in range(50):
  # 5% of young individuals who did experiances side effects. 
  random_young = randint(13,64)
  train_samples.append(random_young)
  train_labels.append(1)

  # 5% of old individuals who did not experiances side effects. 
  random_old = randint(65,100)
  train_samples.append(random_old)
  train_labels.append(0)

for i in range(1000):
  # 95% of young individual wo did not experiance side effects
  random_young = randint(13,64)
  train_samples.append(random_young)
  train_labels.append(0)

  # 95% of old individuals who does experiance side effects
  random_old = randint(65,100)
  train_samples.append(random_old)
  train_labels.append(1)
  

print("End - Assigning age and labels.")
```

```
# for i in train_samples:
#   print(i)

print("Train samples : ", train_samples)
print("Train labels : ", train_labels)

# for i, j in train_samples, train_labels:
#   print("i : ", i, " j : ", j)
```

```
print("Type of the train_labels : ", type(train_labels))
print("Type of the train_samples : ", type(train_samples))

train_labels = np.array(train_labels)
train_samples = np.array(train_samples)

print("Type of the train_labels : ", type(train_labels))
print("Type of the train_samples : ", type(train_samples))

train_labels, train_samples = shuffle(train_labels, train_samples)

print("Done - ")
```

```
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
print("End")
```

```
for i in scaled_train_samples:
  print(i)

# print("Scaled train samples : ", scaled_train_samples)
```

# 4 . Create an Artificial Neural Network with TensorFlow's Keras API


## Simple tf.keras Sequential model

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy

print("End - Import")
```

```
model = Sequential([
   Dense(units=16, input_shape=(1,), activation='relu'),
   Dense(units=32, activation='relu'),
   Dense(units=2, activation='softmax')
])

print("End - creating a model")
```

```
model.summary()
```

# 5.Train an Artificial Neural Network with TensorFlow's Keras API


## compile

```
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print("End - Compile - Optimizer, loss, metrics")
```

## Training

```
model.fit(x=scaled_train_samples, y=train_labels, batch_size=10, epochs=30, shuffle=True, verbose=2)
print("End - fit - x, y, batch_size, epoch, shuffle, verbose")
```

# 6 . Build a Validation Set With TensorFlow's Keras API

```
model.fit(x=scaled_train_samples, y=train_labels, validation_split=0.1, batch_size=10, epochs=30, shuffle=True, verbose=2)
print("End - Adding validation params. It will be a number between 0 to 1, split occurs before the data shuffles ")
```

# 8 . Create a Confusion Matrix for NN predictions

```
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
print("Import done")
```

```
cm = confusion_matrix(y_true=test_labels, y_pred=rounded_predictions)
print("Confusion Matrix - done")
```

```
def plot_confusion_matrix(cm, classes,
                        normalize=False,
                        title='Confusion matrix',
                        cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```
cm_plot_labels = ["No_side effets", "Had_side effects"]
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title="Confusion Matrix")
```
